Notebook ini adalah catatan perjalanan eksplorasi mandiri selama satu semester. Setiap entri diisi setelah menyelesaikan Komponen Mandiri di bab terkait, *sebelum* sesi presentasi dimulai.

**Cara menggunakan notebook ini:**
- Isi satu entri per pekan; jangan menunda.
- Pilih satu pertanyaan atau topik dari materi minggu itu - bebas, tidak harus dari pertanyaan pemantik di bab.
- Bagian *Koneksi ke Pekan Sebelumnya* adalah bagian terpenting - ia memaksa Anda melihat pola dalam belajar Anda sendiri.
- Jangan menghapus atau mempercantik entri lama. Entri jujur tentang kegagalan lebih bernilai dari entri yang dibuat-buat bagus.
- Pada akhir W10, isi sel *Refleksi Portofolio* di bagian bawah.

Template entri lengkap dan panduan presentasi 10 menit ada di `14_Lampiran.md` bagian C.6 dan C.7.
